Q4. Indexing using .loc

Using .loc, perform the following:

1. Display the row with index 10.
2. Display the Sales value of the row with index 20


```python
print(df.loc[10])
print()
print(df.loc[20, "Sales"])
```

Q5. Indexing using .iloc

Using .iloc, display:

Q7. Filtering Data

Filter the dataset to display:

1. Rows where Region = "East".
2. Rows where Sales > 1500.


```python
print(df[df["Region"] == "East"])
print()
print(df[df["Sales"] > 1500])
```

Q8. Multiple Condition Filtering

Display rows where:

- Region is West
- Sales is greater than 1200


```python
print(df[(df["Region"] == "West") & (df["Sales"] > 1200)])
```

Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

- Mean sales value
- Maximum quantity sold
- Minimum sales value


```python
desc = df.describe()
print(desc)
print()
print("Mean sales value:", desc.loc["mean", "Sales"])
print("Maximum quantity sold:", desc.loc["max", "Quantity"])
print("Minimum sales value:", desc.loc["min", "Sales"])
```

Q10. Aggregation Analysis

Using the Sales column, calculate:

1. Total sales
2. Average sales
3. Maximum sales
4. Minimum sales


```python
sales = df["Sales"]

print("Total sales:", sales.sum())
print("Average sales:", sales.mean())
print("Maximum sales:", sales.max())
print("Minimum sales:", sales.min())
```
